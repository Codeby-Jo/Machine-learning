```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
DT = pd.read_csv("study_hours.csv")
DT.head()
```

```python
#load dataset
x = DT.iloc[:,0:1].values
y = DT.iloc[:,1].values
```

```python
#from linear Regression

from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x,y)
y_lin = lin_reg.predict(x)
```

```python
# polynomial Regression
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree = 2) # quardratic curve
x_poly = poly.fit_transform(x)
```

```python
poly_reg = LinearRegression()
poly_reg.fit(x_poly,y)
y_poly = poly_reg.predict(x_poly)
```

```python
# Plot Comparison 
plt.scatter(x,y)
plt.plot(x,y_lin)
plt.plot(x, y_poly)
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.title("Linear vs Polynomial Regression")
plt.show()
```

```python
# -------- Compare R2 Scores --------
from sklearn.metrics import r2_score
print("Linear R2 Score:", r2_score(y, y_lin))
print("Polynomial R2 Score:", r2_score(y, y_poly))
```

```python
import joblib 
joblib.dump(poly, "Polynomial_transform.pkl")
joblib.dump(poly_reg, "poly_regression.pkl")
print("success")
```
